## Problema con las listas: muy intensivas en memoria

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
output = [analyzer.polarity_scores(tweet)['compound'] for tweet in df['tweet']]
output
```

En principio, esta poderÃ­a parecer una manera muy *pythonica* de resolver nuestro problema de analizar la polaridad de todos y cada uno de los tweets. Sin embargo, tiene un **problema** muy importante que es el **uso de memoria**. En este ejemplo, no sucede nada porque estamos analizando muy pocos tweets, pero con millones de ellos, una estructura de tipo lista, que almacena todo en memoria, nos darÃ­a un error de *Out of Memory*.

## SoluciÃ³n 1: Ir escribiendo en el DF

Una posible soluciÃ³n, para no tener que almacenar todos los tweets en memoria y pasarle el analizador de Vader, es ir leyendo el DataFrame de Pandas fila a fila y escribiendo el valor del compound o lo que nos interese en una nueva columna. No es la mÃ¡s ortodoxa ni la mÃ¡s elegante pero funciona.

```python
df["sent"] = [None] * len(df) ## Creamos una columna llena de Nones que es donde vamos a ir escribiendo
df
```

```python
count = 0
for tweet in df.tweet:
    compound = analyzer.polarity_scores(tweet)['compound']
    df.iloc[count,5] = compound ## 5 es el nÃºmero de columna de sent
    count +=1
```

```python
df
```

## SoluciÃ³n 2: Usar un generador en vez de una lista

PodÃ©is utilizar la notaciÃ³n Pythonica de nuevo y utilizar un generador en vez de una lista. De este modo, los datos ya no estarÃ¡n en memoria, si no que estarÃ¡n en disco hasta que los llamÃ©is.

```python
output = (analyzer.polarity_scores(tweet)['compound'] for tweet in df['tweet'])
output
```

**Importante**: no hay una manera directa de asignar un generador a una columna de pandas, pero podemos hacer algo asÃ­:

```python
df["sent"] = output ### Esto no se puede hacer de manera directa, si fuese una lista SÃ
```

```python
for tweet, compound in zip(df.tweet, output):
    print(tweet, "-->", compound)
```

**Importante**: los generadores se vacÃ­an una vez que los recorres una vez:

```python
for compound in output:
    print(compound)
```

## SoluciÃ³n 3: Usar una funciÃ³n lambda

La soluciÃ³n mÃ¡s elegante, pero que en alguna ocasiÃ³n puede llegar a dar problemas de memoria. Utilizamos una funciÃ³n lambda para ejecutar sobre cada uno de nuestros tweets el analyzer de Vader.

```python
df["sent"] = df["tweet"].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df
```
